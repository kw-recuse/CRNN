### Upload the model to HF

```python
import torch
from huggingface_hub import HfApi

checkpoint_path = "/content/checkpoints/epoch1_step37628_loss0.0902.pt"
checkpoint = torch.load(checkpoint_path, map_location="cpu")
trainer.model.load_state_dict(checkpoint["model_state_dict"])
torch.save(trainer.model.state_dict(), "model_weights.pth")
```

```python
from huggingface_hub import HfApi

api = HfApi()
repo_id = "recuse/CRNN_Synth"
api.create_repo(repo_id=repo_id, repo_type="model", exist_ok=True)
```

```python
api.upload_file(
    path_or_fileobj="model_weights.pth",
    path_in_repo="model_weights.pth",
    repo_id=repo_id,
    repo_type="model"
)
```
